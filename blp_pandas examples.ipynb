# blp_pandas Examples

By Teddy Ambona

This pandas wrapper module simplifies the interactions with the Bloomberg API

All the functions return a pandas dataframe indexed on time and tickers

```python
from blp_pandas import blp_pandas as bbg
from datetime import datetime
```

# Intraday Tick Data

```python
# Parameters
security = ['CAC FP Equity'] # A list/string of securities can be passed as argument 
fields = ['TRADE','BID', 'ASK']
start_date = datetime(2018,11,12,10,0)
end_date = datetime(2018,11,12,11,0)
other_param = {'includeNativeTradeId': 'True'}

df = bbg.IntradayTick(security,
                          fields,
                          start_date,
                          end_date,
                          condition_codes=True,
                          other_param=other_param)

df.head()
```

# Intraday Bar Data

```python
# Parameters
security = ['CAC FP Equity'] # A list/string of securities can be passed as argument 
fields = 'BID'
start_date = datetime(2018,11,12,10,0)
end_date = datetime(2018,11,12,11,0)
other_param = None

df = bbg.IntradayBar(security,
                     fields,
                     start_date,
                     end_date,
                     barInterval=10,
                     other_param=other_param)

df.head()
```

# Reference Data

```python
# Parameters
security = ['CAC FP Equity', 'AAPL IM Equity'] # A list/string of securities can be passed as argument 
fields = ['ID_ISIN', 'AVERAGE_CLOSE_AUCTION_VOLUME']
start_date = datetime(2018,11,12,10,0)
end_date = datetime(2018,11,12,11,0)
overrides = {'CALC_INTERVAL': '10D'}
other_param = None

df = bbg.RefData(security,
                 fields,
                 overrides=overrides,
                 other_param=other_param)

df.head()
```

# Historical Data

```python
# Parameters
security = ['CAC FP Equity'] # A list/string of securities can be passed as argument 
fields = ['PX_LAST', 'VOLUME']
start_date = datetime(2018,1,1)
end_date = datetime(2018,11,20)
overrides = None
other_param = None

df = bbg.HistoData(security,
                   fields,
                   start_date,
                   end_date,
                   overrides=overrides,
                   other_param=other_param)

df.head()
```
